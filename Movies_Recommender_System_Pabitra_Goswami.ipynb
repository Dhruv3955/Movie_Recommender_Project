```python
import numpy as mp
import pandas as pd
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tmdb_5000_movies (1).csv')
credits = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tmdb_5000_credits (1).csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#Merging
movies = movies.merge(credits,on ='title')
movies.head(1)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#Now the final DATA that we will be working on
movies.head()
```

```python
#Now we have to do Preprocessing to find missing data......
movies.isnull().sum()
```

```python
#we have missing data in overview which is 3, we have to remove it
movies.dropna(inplace=True)
```

```python
#to find Duplicate data
movies.duplicated().sum()
```

```python
#Now we are going to fix every column of the DATA
#first we have to convert string to list to list
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
#Now one column is fixed which is genres
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
      if counter !=3:
          L.append(i['name'])
          counter+=1
      else:
          break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#now we searching for director in Crew
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
       if i['job']=='Director':
           L.append(i['name'])
           break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#we have to change sam mendes to sammendes
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#now will make a tag collumn where we are going to concadinate(add) all the other columns
movies['tags'] = movies['overview'] + movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#NOW THE FINAL FRESH DATA IS HERE
new_df = movies[['movie_id','title','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
      y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#we are going to make it lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
#now all the data....1st PHASE IS COMPLETE
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
#we have to find similarity between tags
#Text(tags) to VECTOR
#bag of words technique - Vectorization(stop word not allowed)
# tag1+tag2+....

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]

  for i in movies_list:
      print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
